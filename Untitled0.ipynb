```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'news_sold': [2800, 3100, 2900, 3300, 3400, 2500, 2700, 3600, 3900, 2200, 3000, 4100]
}

df = pd.DataFrame(data)
```

```python
#1.Using matplotlib, plot a line graph showing the number of news sold for each month
plt.plot(df['month'], df['news_sold'], marker='o', linestyle='-', color='blue')
plt.title('News Sold per Month')
plt.xlabel('Month')
plt.ylabel('Number of News Sold')
plt.grid(True)
plt.show()
```

# New Section

```python
#2.Add a new column to the DataFrame called "sales_category" that labels months as "High" if news_sold is above 3000 and "Low" otherwise
df['sales_category'] = df['news_sold'].apply(lambda x: 'High' if x > 3000 else 'Low')
```

```python
#3.Sort the DataFrame in descending order based on the news_sold column and display the result
sorted_df = df.sort_values(by='news_sold', ascending=False)
print(sorted_df)
```

```python
#4.Calculate and print the average number of news sold
average_news = df['news_sold'].mean()
print("Average number of news sold:", average_news)
```

```python
#5.Find and display the month(s) with the maximum number of news sold

max_sold = df['news_sold'].max()
max_months = df[df['news_sold'] == max_sold]['month'].tolist()
print("Month(s) with maximum news sold:", max_months)
```
